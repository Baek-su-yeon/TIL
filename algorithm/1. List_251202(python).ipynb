```python
# 16118. 구간합
T = int(input()) # 테스트 케이스 개수
for test_case in range(1, T + 1):
    n, m = map(int, input().split()) # 정수 개수, 구간 개수
    num_list = list(map(int, input().split()))

    max_sum = 0
    min_sum = 10000 * 99 # 문제 제한사항: 10 <= n <= 100, 2 <= m < n, 1<=num<=10000
    for i in range(n - m + 1):
        m_list = num_list[i:i+m]
		
        m_sum = 0
        for num in m_list:
            m_sum += num
        
        if max_sum < m_sum:
            max_sum = m_sum
        if m_sum < min_sum:
            min_sum = m_sum

    print(f'#{test_case} {max_sum - min_sum}')
```

```python
# 1206. [S/W 문제해결 기본] 1일차 - View
# 고쳐라
T = 10
VISIBLE_DIS = 2

for test_case in range(1, T + 1):
    n = int(input()) # 건물 개수 (4 <= n <= 1000)
    heights = list(map(int, input().split())) # 건물 높이 (0 <= height <= 255)
    visible = 0 # 조망권이 있는 세대 개수

    for i in range(n):
        if not heights[i]: # 높이 = 0이면 건물 없으므로 pass
            continue

        center = heights[i]
        near_height = 0
        for j in range(1, VISIBLE_DIS + 1):
            if center < heights[i - j] or center < heights[i + j]:
                break
            
            max(near_height, heights[i - j], heights[i + j])
        else:
            visible += center - near_height

    print(f'#{test_case} {visible}')
```
