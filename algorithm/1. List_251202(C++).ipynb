```C++
// C++
// 16118. 구간합
#include <iostream>
#include <vector>
#include <algorithm> // std::max, std::min 사용

using namespace std;

int main()
{
    int T = 0;
    cin >> T;

    for (int i = 1; i < T + 1; i++)
    {
        int n, m;
        cin >> n >> m;

        vector<int> num_list(n);
        for (int i = 0; i < n ; i++)
        {
            cin>>num_list[i];
        }

        
        int max_sum = 0;
        int min_sum = 10000*99;
        int now_sum = 0;

        // 초기 윈도우 설정
        for (int i = 0; i < m; i++)
        {
            now_sum += num_list[i]
        }

        max_sum = now_sum
        min_sum = now_sum

        // 윈도우 슬라이딩 for문 두개 보다 빠르고 효과적(O(N))
        for (int i = 1; i < n - m + 1; i++)
        {
            now_sum -= num_list[i - 1];
            now_sum += num_list[i + m - 1];

            if (max_sum < now_sum) max_sum = now_sum;
            if (now_sum < min_sum) min_sum = now_sum;
        }
        cout << "#" << test_case << " " << max_sum - min_sum << endl;
    }

    return 0;
}
```
