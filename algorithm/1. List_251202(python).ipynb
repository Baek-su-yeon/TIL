```python
# 1206. [S/W 문제해결 기본] 1일차 - View
# 사실 max() 쓰면 nears 만들 필요 없음
T = 10
for test_case in range(1, T + 1):
    n = int(input()) # 건물 개수
    heights = list(map(int, input().split())) # 건물 높이
    result = 0

    for i in range(n):
        # 맨 왼쪽 두칸, 맨 오른쪽 두칸 제외
        if heights[i] == 0:
            continue

        center = heights[i]
        nears =  [
            heights[i - 2], 
            heights[i - 1],
            heights[i + 1],
            heights[i + 2]
        ]

        max_near = 0
        for near in nears:
            if max_near < near:
                max_near = near

        if max_near < center:
            result += center - max_near
    
    print(f'#{test_case} {result}')
```

```python
# 24735. Gravity
# 낙하할 수 있는 공간 = 오른쪽 벽까지 공간 중 다른 상자에 의해 가로 막히지 않는 공간
# 가로 막히지 않는 공간 = 상자 높이 보다 낮은 높이의 상자 개수

T = 1
for test_case in range(1, T + 1):
    n = int(input()) # 박스가 높인 가로 칸의 수
    boxes = list(map(int, input().split())) # 각 칸의 상자 높이

    max_overlap = 0
    for i in range(n):
        now_box = boxes[i]
        overlap = 0
        for j in range(i, n):
            if boxes[j] < now_box:
                overlap += 1

        if max_overlap < overlap:
            max_overlap = overlap
    
    print(max_overlap)
```

```python
# 24760. Baby Gin
# run: 3장의 카드가 연속적인 번호를 갖는 경우
# triplet: 3장의 카드가 동일한 번호를 갖는 경우
# baby-gin: 6장의 카드가 run과 triplet으로 구성된 경우

T = int(input())
CARD_NUM = 10 # 0 ~ 9 사이의 카드

for test_case in range(1, T + 1):
    cards = input().strip()
    card_count = [0] * CARD_NUM

    # 어떤 카드 있는지 탐색
    for card in cards:
        card_count[int(card)] += 1

    idx = 0
    triplet = 0
    card_run = 0 # 해당 문제 변수명 run 사용 불가

    # 한 카드에서 모든 경우의 수를 다 찾아야함 주의!!!!!!!!!!!
    # 예) 1 2 3 1 2 3
    while idx < CARD_NUM:

        # triplet 할 수 있는 상황
        if card_count[idx] >= 3: 
            triplet += 1
            card_count[idx]-= 3
            continue

        # run할 수 있는 상황
        if idx + 2 < 10 and card_count[idx] >= 1 and card_count[idx + 1] >= 1 and card_count[idx + 2] >= 1:
            card_run += 1
            card_count[idx] -= 1
            card_count[idx + 1] -= 1
            card_count[idx + 2] -= 1
            continue
        
        idx += 1
    
    if triplet + card_run == 2:
        print(f'#{test_case} Baby Gin')
    else:
        print(f'#{test_case} Lose')
```

```python
# 16192. 전기버스
# 최소한 몇 번의 충전을 해야 종점에 도착할 수 있는가?
# 출발지에는 항상 충전기가 설치되어 있지만 충전횟수에는 포함하지 않는다

T = int(input())
for test_case in range(1, T + 1):
    k,n,m = map(int, input().split()) # k: 이동거리, n: 종점, m: 충전소 개수
    chargers = list(map(int, input().split()))
    road = [0] * (n + 1)

    # 충전기 있는 곳 표시
    for charger in chargers:
        road[charger] = 1

    result = 0
    now = 0
    while True:

        if now + k >= n:
            break
        
        if road[now + k]: # 충전기가 있다
            now += k
            result += 1

        else: # 충전기가 없다
            for i in range(now + k - 1, now, -1): # 충전소 있는 곳 찾기
                if road[i]:
                    now = i
                    result += 1
                    break
            else:
                result = 0
                break
    
    print(f'#{test_case} {result}')
```

```python
# 간단한 소인수분해

T = int(input())
for test_case in range(1, T + 1):
    num = int(input())
    base = [2,3,5,7,11]
    exp = [0] * len(base)

    # 지수 계산
    for i in range(len(base)):

        while num % base[i] == 0:
            exp[i] += 1
            num //= base[i]
    
    # 지수 출력
    print(f'#{test_case}', end=' ')
    for num in exp:
        print(num, end=' ')
    print()
```

```python
# 12712. 파리퇴치3

T = int(input())
for test_case in range(1, T + 1):
    n, m = map(int, input().split()) # n: 배열 크기, m: 분사 거리
    fly = [list(map(int, input().split())) for _ in range(n)]

    plus_shape = [(-1, 0), (1, 0), (0, -1), (0, 1)] # 상하좌우
    x_shape= [(-1, -1), (-1, 1), (1, -1), (1, 1)]

    max_die = 0
    for y in range(n):
        for x in range(n):
            plus_die = fly[y][x]
            x_die = fly[y][x]

            # 죽은 파리 세기
            for dis in range(1, m):
                # plus
                for dy, dx in plus_shape:
                    ny = y + (dy * dis)
                    nx = x + (dx * dis)
                    if 0 <= ny < n and 0 <= nx < n:
                        plus_die += fly[ny][nx]

                # x
                for dy, dx in x_shape:
                    ny = y + (dy * dis)
                    nx = x + (dx * dis)
                    if 0 <= ny < n and 0 <= nx < n:
                        x_die += fly[ny][nx]
            
            cur_die = max(plus_die, x_die)
            max_die = max(max_die, cur_die)
    
    print(f'#{test_case} {max_die}')
```

```python
# 1979. 어디에 단어가 들어갈 수 있을까

T = int(input())
for test_case in range(1, T + 1):
    n, k = map(int, input().split()) # n: 퍼즐 크기, k: 단어 길이
    puzzle = [list(map(int, input().split())) for _ in range(n)] # 1:흰색(빈곳), 0:검은색(벽)
    result = 0

    # 횡 순회
    for i in range(n):
        blank = 0
        for j in range(n):
            if puzzle[i][j] == 1:
                blank += 1
            else: # 벽 만남
                if blank == k:
                    result += 1
                blank = 0 # 초기화
        
        if blank == k:
            result += 1
    
    # 열 순회
    for j in range(n):
        blank = 0 
        for i in range(n):
            if puzzle[i][j] == 1:
                blank += 1
            else:
                if blank == k:
                    result += 1
                blank = 0
        
        if blank == k:
            result += 1

    print(f'#{test_case} {result}')
```

```python
# 1974. 스토쿠 검증

PUZZLE_SIZE = 9

# 횡 검사
def row_check(arr):
    for i in range(PUZZLE_SIZE):
        num_count = [0] * (PUZZLE_SIZE + 1)
        for j in range(PUZZLE_SIZE):
            num = arr[i][j]
            num_count[num] += 1
            
            if num_count[num] > 1:
                return False
    return True

# 열 검사
def col_check(arr):
    for j in range(PUZZLE_SIZE):
        num_count = [0] * (PUZZLE_SIZE + 1)
        for i in range(PUZZLE_SIZE):
            num = arr[i][j]
            num_count[num] += 1

            if num_count[num] > 1:
                return False
    return True

# 사각형 검사
def squre_check(arr):
    for i in range(0, PUZZLE_SIZE, 3):
        for j in range(0, PUZZLE_SIZE, 3):
            num_count = [0] * (PUZZLE_SIZE + 1)
            for k in range(3):
                for l in range(3):
                    num = arr[i + k][j + l]
                    num_count[num] += 1

                    if num_count[num] > 1:
                        return False
    return True

T = int(input())
for test_case in range(1, T + 1):
    puzzle = [list(map(int, input().split())) for _ in range(PUZZLE_SIZE)]
    
    if row_check(puzzle) and col_check(puzzle) and squre_check(puzzle):
        print(f'#{test_case} 1')
    else:
        print(f'#{test_case} 0')
```

```python
# 1209. Sum

ARRAY_SIZE = 100

def row_sum(arr):
    sums = []
    for i in range(ARRAY_SIZE):
        sums.append(sum(arr[i]))
    
    return max(sums)

def col_sum(arr):
    sums = []
    for j in range(ARRAY_SIZE):
        col_sum = 0
        for i in range(ARRAY_SIZE):
            col_sum += arr[i][j]
        sums.append(col_sum)

    return max(sums)

def diagonal_sum(arr):
    right_dia = 0
    left_dia = 0

    for i in range(ARRAY_SIZE):
        right_dia += arr[i][i]
        left_dia += arr[i][ARRAY_SIZE - i - 1]
    
    return max(right_dia, left_dia)

T = 10
for test_case in range(1, T + 1):
    case = int(input())
    arr = [list(map(int, input().split())) for _ in range(ARRAY_SIZE)]

    max_row = row_sum(arr)
    max_col = col_sum(arr)
    max_dia = diagonal_sum(arr)

    print(f'#{test_case} {max(max_row, max_col, max_dia)}')
```
