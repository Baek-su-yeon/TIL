```python
# 16478. 반복문자 지우기

T = int(input())
for test_case in range(1, T + 1):
    str_word = input()

    stack = []
    for word in str_word:
        
        if not stack:
            stack.append(word)
            continue

        if stack[-1] == word:
            stack.pop()
        else:
            stack.append(word)
    
    print(f'#{test_case} {len(stack)}')
```

```python
# 16476. 괄호검사

T = int(input())
for test_case in range(1, T + 1):
    string = input()
    stack = []

    result = 1
    for word in string:

        if word in '{[(':
            stack.append(word)
        
        elif word in '}])':
            if not stack:
                result = 0
                break
            elif stack[-1] == '{' and word == '}' or stack[-1] == '[' and word == ']' or stack[-1] == '(' and word == ')':
                stack.pop()
            else:
                result = 0
                break
    
    # 스택이 비었다.
    if stack:
        result = 0

    print(f'#{test_case} {result}')
```

```python
# 2005. 파스칼의 삼각형

T = int(input())
for test_case in range(1, T + 1):
    n = int(input()) # 삼각형 크기
    pascal = [[0] * n for _ in range(n)]

    print(f'#{test_case}')
    
    # 파스칼 점화식 -> p[i][j] = p[i-1][j-1] + p[i-1][j]
    for i in range(n):
        for j in range(i + 1):

            if j == 0 or j == i: # 양쪽 끝에 있는 상황
                pascal[i][j] = 1
                print(pascal[i][j], end=' ')
            
            else:
                pascal[i][j] = pascal[i - 1][j - 1] + pascal[i - 1][j]
                print(pascal[i][j], end=' ')
        
        print()
```

```python
# 16506. 그래프 경로

def DFS(adj, v, start, goal):
    visited = [0] * (v + 1)
    stack = []

    # 시작점 처리
    stack.append(start)
    visited[start] = 1

    while stack:
        # 현재 위치는?
        now = stack.pop()
        if now == goal: # 지금 위치가 목적지야?
            return 1

        # 인접한 노드는?
        for node in adj[now]:
            if not visited[node]:
                stack.append(node)
                visited[node] = 1
    
    return 0

T = int(input())
for test_case in range(1, T + 1):
    v, e = map(int, input().split()) # v: 노드 수, e: 간선 수
    adj_list = [[] for _ in range(v + 1)] # 기억해! 빈 공간을 !!! 늘려야한다!!!!!!!!!!!!!

    # 방향성 그래프(단방향)
    for _ in range(e):
        start, end = map(int, input().split())
        adj_list[start].append(end)

    s, g = map(int, input().split())

    result = DFS(adj_list, v, s, g)

    print(f'#{test_case} {result}')
```

```python
# 16505. 종이붙이기

def paper(n, memo):
    # 20 * 10k 크기를 덮는 방법의 수
    # 20 * 20을 쓰는거는 20 * 10 2개를 가로/세로로 놓는 2개의 경우의 수를 가짐
    # 결국 전 단계, 전 전단계에서 10을 붙이냐, 20을 붙이냐
    # A2 = A1 + 2*A0 = 1 + 2 = 3
    # A3 = A2 + 2*A1 = 3 + 2 = 5
    # Ak = A(k-1) + 2*A(k-2)
    k = n // 10
    for i in range(2, k + 1):
        memo[i] = memo[i - 1] + 2 * memo[i - 2]
    
    return memo[k]


T = int(input())
for test_case in range(1, T + 1):
    n = int(input())

    memo = [0] * 31 # 0 <= n <= 300, n: 10의 배수, 인덱싱으로 기억 즉 1이 10일때
    memo[0] = 1 # 아무 것도 없는 경우, A0
    memo[1] = 1 # 20 * 10 하나만 쓰는 경우, A1

    result = paper(n, memo)
    print(f'#{test_case} {result}')
```
