```python
# 16118. 구간합
T = int(input()) # 테스트 케이스 개수
for test_case in range(1, T + 1):
    n, m = map(int, input().split()) # 정수 개수, 구간 개수
    num_list = list(map(int, input().split()))

    max_sum = 0
    min_sum = 10000 * 99 # 문제 제한사항: 10 <= n <= 100, 2 <= m < n, 1<=num<=10000
    for i in range(n - m + 1):
        m_list = num_list[i:i+m]
		
        m_sum = 0
        for num in m_list:
            m_sum += num
        
        if max_sum < m_sum:
            max_sum = m_sum
        if m_sum < min_sum:
            min_sum = m_sum

    print(f'#{test_case} {max_sum - min_sum}')
```

```python
# 1206. [S/W 문제해결 기본] 1일차 - View
# 사실 max() 쓰면 nears 만들 필요 없음
T = 10
for test_case in range(1, T + 1):
    n = int(input()) # 건물 개수
    heights = list(map(int, input().split())) # 건물 높이
    result = 0

    for i in range(n):
        # 맨 왼쪽 두칸, 맨 오른쪽 두칸 제외
        if heights[i] == 0:
            continue

        center = heights[i]
        nears =  [
            heights[i - 2], 
            heights[i - 1],
            heights[i + 1],
            heights[i + 2]
        ]

        max_near = 0
        for near in nears:
            if max_near < near:
                max_near = near

        if max_near < center:
            result += center - max_near
    
    print(f'#{test_case} {result}')
```

```python
# 24735. Gravity
# 낙하할 수 있는 공간 = 오른쪽 벽까지 공간 중 다른 상자에 의해 가로 막히지 않는 공간
# 가로 막히지 않는 공간 = 상자 높이 보다 낮은 높이의 상자 개수

T = 1
for test_case in range(1, T + 1):
    n = int(input()) # 박스가 높인 가로 칸의 수
    boxes = list(map(int, input().split())) # 각 칸의 상자 높이

    max_overlap = 0
    for i in range(n):
        now_box = boxes[i]
        overlap = 0
        for j in range(i, n):
            if boxes[j] < now_box:
                overlap += 1

        if max_overlap < overlap:
            max_overlap = overlap
    
    print(max_overlap)
```

```python
# 24760. Baby Gin
# run: 3장의 카드가 연속적인 번호를 갖는 경우
# triplet: 3장의 카드가 동일한 번호를 갖는 경우
# baby-gin: 6장의 카드가 run과 triplet으로 구성된 경우

T = int(input())
CARD_NUM = 10 # 0 ~ 9 사이의 카드

for test_case in range(1, T + 1):
    cards = input().strip()
    card_count = [0] * CARD_NUM

    # 어떤 카드 있는지 탐색
    for card in cards:
        card_count[int(card)] += 1

    idx = 0
    triplet = 0
    card_run = 0 # 해당 문제 변수명 run 사용 불가

    # 한 카드에서 모든 경우의 수를 다 찾아야함 주의!!!!!!!!!!!
    # 예) 1 2 3 1 2 3
    while idx < CARD_NUM:

        # triplet 할 수 있는 상황
        if card_count[idx] >= 3: 
            triplet += 1
            card_count[idx]-= 3
            continue

        # run할 수 있는 상황
        if idx + 2 < 10 and card_count[idx] >= 1 and card_count[idx + 1] >= 1 and card_count[idx + 2] >= 1:
            card_run += 1
            card_count[idx] -= 1
            card_count[idx + 1] -= 1
            card_count[idx + 2] -= 1
            continue
        
        idx += 1
    
    if triplet + card_run == 2:
        print(f'#{test_case} Baby Gin')
    else:
        print(f'#{test_case} Lose')
```

```python
# 16192. 전기버스
# 최소한 몇 번의 충전을 해야 종점에 도착할 수 있는가?
# 출발지에는 항상 충전기가 설치되어 있지만 충전횟수에는 포함하지 않는다

T = int(input())
for test_case in range(1, T + 1):
    k,n,m = map(int, input().split()) # k: 이동거리, n: 종점, m: 충전소 개수
    chargers = list(map(int, input().split()))
    road = [0] * (n + 1)

    # 충전기 있는 곳 표시
    for charger in chargers:
        road[charger] = 1

    result = 0
    now = 0
    while True:

        if now + k >= n:
            break
        
        if road[now + k]: # 충전기가 있다
            now += k
            result += 1

        else: # 충전기가 없다
            for i in range(now + k - 1, now, -1): # 충전소 있는 곳 찾기
                if road[i]:
                    now = i
                    result += 1
                    break
            else:
                result = 0
                break
    
    print(f'#{test_case} {result}')
```

```python
# 16191. 숫자카드

CARD_NUM = 10

T = int(input())
for test_case in range(1, T + 1):
    n = int(input()) # 카드 장수
    cards = input()

    card_count = [0] * CARD_NUM # 0~9까지 카드 존재
    for card in cards:
        card_count[int(card)] += 1
    
    max_num = 0
    max_card = 0
    for i in range(CARD_NUM):
        if max_num <= card_count[i]: # 같을 때는 숫자가 큰 수를 출력해야함
            max_card = i
            max_num = card_count[i]
    
    print(f'#{test_case} {max_card} {max_num}')
```
